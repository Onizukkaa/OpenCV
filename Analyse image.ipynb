```python
def border(image,k):
    im_temp=image.copy()
    im_temp[:,:k]=0
    im_temp[:,-k:]=0
    im_temp[:k,:]=0
    im_temp[-k:,:]=0
    plt.imshow(im_temp)
border(damier,2)
```

```python
def reverse(im):
    im_temp=im.copy()
    im_temp=255-im_temp
    plt.imshow(im_temp)
    
reverse(damier)
```

```python
def masque(im,h,i,l,j):
    im_temp=im.copy()
    im_temp[h:i,l:j]=255
    plt.imshow(im_temp)
    
masque(damier,70,100,75,150)
```

```python
def inverse_image(im):
    im_temp=im.copy()
    
    im_temp=im_temp[::-1,:,:]
    
    plt.imshow(im_temp)
inverse_image(damier)   
```

```python
def inverse_image(im):
    im_temp=im.copy()
    
    im_temp=im_temp[:,::-1,:]
    
    plt.imshow(im_temp)
inverse_image(damier)   
```

```python
K=np.array([[1,2,4,2,1],[2,4,8,4,2],[4,8,16,8,4],[2,4,8,4,2],[1,2,4,2,1]])/100.0
print(K)

def filter(im,K):
    im_temp=im.copy()*0
    s=int((K.shape[0]+1)/2)
    for i in range(s,im.shape[0]-s):
        for j in range(s,im.shape[1]-s):
            im_temp[i,j]=np.sum(K*im[i-(s-1):i+s,j-(s-1):j+s])
    return(im_temp)

plt.imshow(filter(damier[:,:,0],K),cmap="gray")
```
