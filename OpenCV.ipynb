# Dssiner des formes gomtriques

```python
def affichage(i):
    plt.imshow(i[:,:,::-1])#remet les couleurs dans l'ordre
    plt.show()
```

# dessiner lignes

```python
img=cv2.line(img,(0,0),(0,400),(255,0,0),20)
affichage(img)
```

```python
pts=np.array([[0,H-1],[W-1,H-1],[W-1,0]], np.int32)
img=cv2.polylines(img,[pts],False,(0,255,0),20)
affichage(img)
```

# Flches

```python
img=cv2.arrowedLine(img,(0,0),(200,400),(255,255,0),20)
affichage(img)
```

# Rectangle

```python
img=cv2.rectangle(img,(200,700),(900,20),(0,25)5,0,20)
affichage(img)
```
